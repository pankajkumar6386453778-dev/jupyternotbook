```python
import matplotlib.pyplot as plt

#data
months = ["jan", "feb","mar","apr" , "may"]
sales =[120,150,180,170,200]
profit =[20,35,40,30,50]
```

```python
#figure
plt.figure(figsize=(12,8))
```

```python
#line plot
plt.subplot(2,2,1)
plt.plot(months , sales,
         color ="blue",
         marker = "o",
         linestyle = "-",
         linewidth =2,
         label = "sales")

plt.plot(months ,profit,
         color ="red",
         marker = "s",
         linestyle = "--",
         linewidth =2,
         label = "Profit")

plt.title("Sales vs Profit")
plt.xlabel("Months")
plt.ylabel("Amount")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#bar plot
plt.subplot(2,2,2)

plt.bar(months , sales , color = "green")

plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
            
```

```python
#scatter plot
plt.subplot(2,2,3)

plt.scatter(sales , profit,
            color="purple",
            s = 120,
            marker ="*")

```
