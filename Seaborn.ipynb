```python
#Load datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
penguins = sns.load_dataset("penguins")
```

```python
#Theme and styles
sns.set_theme(style="darkgrid")
sns.set_theme(style="white")
sns.set_theme(style="ticks")
sns.set_theme(style="dark")
sns.set_theme(style="whitegrid")
```

```python
#Scatter plot
sns.scatterplot(
    data = tips,
    x = "total_bill",
    y = "tip"
)
plt.show()
```

```python
#Using hue
sns.scatterplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    hue ="sex"
)
```

```python
#Using style
sns.scatterplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    style = "time"
)
```

```python
#using size
sns.scatterplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    size = "time",
    color ="red"
)
```

```python
#Line plot
sns.lineplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    color ="green"
)
plt.show()
```

```python
#Histogram
sns.histplot(
    data = tips,
    x = "total_bill",
    color ="red"
    

)
plt.show()
```

```python
#KDE plot
sns.kdeplot(
    data = tips,
    x = "total_bill",
    fill =True,
    color = "red"

)
plt.show()
```

```python
#KDE 
sns.histplot(
    data = tips,
    x = "total_bill",
    kde =True,
    color = "green"

)
plt.show()
```

```python
#Count plot
sns.countplot(
    data = tips,
    x = "day",
    color = "red"
)
```

```python
#Bar plot
sns.barplot(
    data = tips,
    x = "day",
    y = "total_bill",
    color ="violet"

)
plt.show()
```

```python
#box plot
sns.boxplot(
    data = tips,
    x ="day",
    y = "total_bill",
    color = "red"
    

)
plt.show()
```

```python
#violin plot
sns.violinplot(
    data = tips,
    x = "day",
    y = "total_bill",

    color = "violet"

)
plt.show()
```

```python
#Strip plot
sns.stripplot(
    data = tips,
    x = "day",
    y = "total_bill",
    color ="red"

)
plt.show()
```

```python
#swarm plot
sns.swarmplot(
    data = tips,
    x = "day",
    y = "total_bill",

    color = "violet"

)
plt.show()
```

```python
# Piar plot
sns.pairplot(
    iris,
    hue = "species"
)
```

```python
#Joint plot
sns.jointplot(
    data = tips,
    x = "total_bill",
    y ="day",
    color = "violet"

)
plt.show()
```

```python
#Heatmap
corr =tips.select_dtypes(include = "number").corr()

sns.heatmap(
    corr,
    annot = True,
    cmap = "coolwarm"
)
```

```python
#cluster map
sns.clustermap(corr)
```

```python
#regression plot
sns.regplot(
    data = tips,
    x= "total_bill",
    y = "tip"

)
plt.show()
```

```python
#lmplot
sns.lmplot(
    data = tips,
    x= "total_bill",
    y = "tip",
    hue = "sex"

)
plt.show()
```

```python
#Facetgrid
g= sns.FacetGrid(tips , col= "time")
g.map(sns.scatterplot , "total_bill" , "tip")
```

```python
#pair grid
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
#catplot
sns.catplot(
    data =tips,
    x = "day",
    y = "tip",
    kind = "box",
    color = "red"
    
)
```

```python
#relplot
sns.relplot(
    data = tips,
    x= "total_bill",
    y= "tip",
    hue = "day"
)
```

```python
#displot
sns.displot(
    data = tips,
    x = "total_bill",
    kde = True,
    color = "red"
)
```

```python
#ECDF plot
sns.ecdfplot(
    data = tips,
    x = "total_bill"
)
```

```python
#Rug plot
sns.rugplot(
    data = tips,
    x = "total_bill"
)
```

```python
#Residual plot
sns.residplot(
    data = tips,
    x = "total_bill",
    y = "tip"
    
)
```

```python
#Color platters
sns.color_palette("deep")
sns.color_palette("pastel")
sns.color_palette("bright")
sns.color_palette("viridis")
sns.color_palette("magma")
sns.color_palette("rocket")
```
